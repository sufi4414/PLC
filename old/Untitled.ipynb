```python
def triangle_wave(freq, duration, sample_rate):
    t = np.linspace(0, duration, int(duration * sample_rate), endpoint=False)
    wave = 2 * np.abs((t * freq) % 1 - 0.5) 
#     wave = t
    return wave

sample_rate = 44100  # Sample rate in Hz
duration = 2 # Duration of the wave in seconds
frequency = 5  # Frequency of the triangle wave in Hz

triangle_waveform = triangle_wave(frequency, duration, sample_rate)

# Plot the triangle wave
plt.plot(triangle_waveform)
plt.title('Triangle Waveform')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

```python

# Define parameters
sample_rate = 44100  # Sample rate in Hz
duration = 1  # Duration of the signal in seconds
initial_frequency = 20  # Initial frequency in Hz
final_frequency = 2  # Final frequency in Hz

# Generate time array
t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)

# Generate sine wave with decreasing frequency
frequency = np.linspace(initial_frequency, final_frequency, len(t))
signal = np.sin(2 * np.pi * frequency * t)

# Plot the signal
plt.plot(t, signal)
plt.title('Continuous Sine Wave with Decreasing Frequency')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```
